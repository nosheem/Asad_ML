```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# Create a dummy dataset
np.random.seed(0)
n_samples = 1000
time = np.arange(n_samples)
cooling_load = np.sin(time * 0.02) + np.random.normal(scale=0.1, size=n_samples)
dhw_load = np.cos(time * 0.02) + np.random.normal(scale=0.1, size=n_samples)
equipment_electric_power = np.sin(time * 0.02 + 1) + np.random.normal(scale=0.1, size=n_samples)
carbon_intensity = np.cos(time * 0.02 + 1) + np.random.normal(scale=0.1, size=n_samples)
solar_generation = np.sin(time * 0.02 + 2) + np.random.normal(scale=0.1, size=n_samples)
```

```python
# Create DataFrames
building_data = pd.DataFrame({
    'Cooling Load': cooling_load,
    'DHW Load': dhw_load,
    'Equipment Electric Power': equipment_electric_power
})
neighborhood_data = pd.DataFrame({
    'Carbon Intensity': carbon_intensity,
    'Solar Generation': solar_generation
})
```

```python
# Scaling the data
scaler_building = MinMaxScaler()
scaled_building_data = scaler_building.fit_transform(building_data)

scaler_neighborhood = MinMaxScaler()
scaled_neighborhood_data = scaler_neighborhood.fit_transform(neighborhood_data)
```

```python
# Create Time Series Generators
n_input = 48
building_generator = TimeseriesGenerator(scaled_building_data, scaled_building_data, length=n_input, batch_size=1)
neighborhood_generator = TimeseriesGenerator(scaled_neighborhood_data, scaled_neighborhood_data, length=n_input, batch_size=1)
```

```python
# 2. Model Design

# LSTM Model for Building-Level Forecasting
model_lstm_building = Sequential([
    LSTM(50, activation='relu', input_shape=(n_input, building_data.shape[1])),
    Dense(building_data.shape[1])
])
model_lstm_building.compile(optimizer='adam', loss='mse')

# RNN Model for Neighborhood-Level Forecasting
model_rnn_neighborhood = Sequential([
    SimpleRNN(50, activation='relu', input_shape=(n_input, neighborhood_data.shape[1])),
    Dense(neighborhood_data.shape[1])
])
model_rnn_neighborhood.compile(optimizer='adam', loss='mse')
```

```python
# Train LSTM model for building-level data
model_lstm_building.fit(building_generator, epochs=10)

# Train RNN model for neighborhood-level data
model_rnn_neighborhood.fit(neighborhood_generator, epochs=10)
```

```python
building_predictions = model_lstm_building.predict(building_generator)
neighborhood_predictions = model_rnn_neighborhood.predict(neighborhood_generator)

print("Building-level predictions (first 5 steps):\n", building_predictions[:5])
print("Neighborhood-level predictions (first 5 steps):\n", neighborhood_predictions[:5])
import matplotlib.pyplot as plt
```

```python
def plot_predictions(original, predicted, title, labels):
    plt.figure(figsize=(14, 7))
    for i in range(original.shape[1]):
        plt.subplot(original.shape[1], 1, i + 1)
        plt.plot(original[:, i], label=f'Original {labels[i]}')
        plt.plot(np.arange(n_input, n_input + len(predicted[:, i])), predicted[:, i], label=f'Predicted {labels[i]}')
        plt.legend()
        plt.title(f'{title} - {labels[i]}')
    plt.tight_layout()
    plt.show()
```

```python
# Plot building-level predictions
plot_predictions(scaled_building_data, building_predictions, 'Building-Level Predictions', building_data.columns)

# Plot neighborhood-level predictions
plot_predictions(scaled_neighborhood_data, neighborhood_predictions, 'Neighborhood-Level Predictions', neighborhood_data.columns)
```

```python
# 4. Visualization Functions

# Function to plot original and predicted data
def plot_predictions(original, predicted, title, labels, n_input):
    plt.figure(figsize=(14, 10))
    for i in range(original.shape[1]):
        plt.subplot(original.shape[1], 1, i + 1)
        plt.plot(original[:, i], label=f'Original {labels[i]}', color='blue')
        plt.plot(np.arange(n_input, n_input + len(predicted[:, i])), predicted[:, i], label=f'Predicted {labels[i]}', color='orange')
        plt.axvspan(n_input, n_input + len(predicted[:, i]), color='orange', alpha=0.1)
        plt.legend()
        plt.xlabel('Time Step')
        plt.ylabel('Normalized Value')
        plt.title(f'{title} - {labels[i]}')
    plt.tight_layout()
    plt.show()

# Function to plot prediction errors
def plot_errors(original, predicted, title, labels, n_input):
    plt.figure(figsize=(14, 10))
    for i in range(original.shape[1]):
        errors = original[n_input:, i] - predicted[:, i]
        plt.subplot(original.shape[1], 1, i + 1)
        plt.plot(errors, label=f'Error in {labels[i]}', color='red')
        plt.legend()
        plt.xlabel('Time Step')
        plt.ylabel('Error')
        plt.title(f'Prediction Error - {title} - {labels[i]}')
    plt.tight_layout()
    plt.show()

# Function to plot distribution comparison
def plot_distribution(original, predicted, title, labels, n_input):
    plt.figure(figsize=(14, 10))
    for i in range(original.shape[1]):
        plt.subplot(original.shape[1], 1, i + 1)
        sns.kdeplot(original[n_input:, i], label=f'Original {labels[i]}', color='blue')
        sns.kdeplot(predicted[:, i], label=f'Predicted {labels[i]}', color='orange')
        plt.legend()
        plt.xlabel('Normalized Value')
        plt.title(f'Distribution Comparison - {title} - {labels[i]}')
    plt.tight_layout()
    plt.show()

# Function to plot heatmap of correlation
def plot_heatmap(original, predicted, title, labels, n_input):
    plt.figure(figsize=(10, 8))
    combined = np.hstack((original[n_input:], predicted))
    columns = [f'Original {label}' for label in labels] + [f'Predicted {label}' for label in labels]
    df_combined = pd.DataFrame(combined, columns=columns)
    corr = df_combined.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlation Heatmap - {title}')
    plt.show()

# 5. Visualization Execution

# Plot building-level predictions
plot_predictions(scaled_building_data, building_predictions, 'Building-Level Predictions', building_data.columns, n_input)

# Plot neighborhood-level predictions
plot_predictions(scaled_neighborhood_data, neighborhood_predictions, 'Neighborhood-Level Predictions', neighborhood_data.columns, n_input)

# Plot building-level prediction errors
plot_errors(scaled_building_data, building_predictions, 'Building-Level Errors', building_data.columns, n_input)

# Plot neighborhood-level prediction errors
plot_errors(scaled_neighborhood_data, neighborhood_predictions, 'Neighborhood-Level Errors', neighborhood_data.columns, n_input)

# Plot building-level distribution comparison
plot_distribution(scaled_building_data, building_predictions, 'Building-Level Distribution', building_data.columns, n_input)

# Plot neighborhood-level distribution comparison
plot_distribution(scaled_neighborhood_data, neighborhood_predictions, 'Neighborhood-Level Distribution', neighborhood_data.columns, n_input)

# Plot building-level heatmap
plot_heatmap(scaled_building_data, building_predictions, 'Building-Level Correlation', building_data.columns, n_input)

# Plot neighborhood-level heatmap
plot_heatmap(scaled_neighborhood_data, neighborhood_predictions, 'Neighborhood-Level Correlation', neighborhood_data.columns, n_input)
```
